```python
import requests
import xmltodict

url = 'http://www.cbr.ru/scripts/XML_daily.asp'
responce = requests.get(url)
responce
```

```python
responce.content
```

```python
data = xmltodict.parse(responce.content)
data
```

```python
items = []
for item in data['ValCurs']['Valute']:
  items.append(item)
items
```

```python
my_array = []
for item in data['ValCurs']['Valute']:
  my_set = [item['NumCode'], item['Name'], item['Value']]
  my_array.append(my_set)
  print(my_set)
```
